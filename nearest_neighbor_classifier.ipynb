```python
path = "/content/drive/MyDrive/cs231n/"
def unpickle(file):
    import pickle
    with open(path + file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
def L1_distance(X1: np.ndarray, X2: np.ndarray) -> float:
  j = np.meshgrid(range(X1.shape[0]))
  return sum(abs(X1[j]-X2[j]))

def min(X: np.ndarray):
  n = X.shape[0]
  result = X[0]
  for i in range(n):
    if result > X[i]:
      result = X[i]
  return result
```

```python
class NearestNeighbor(object):
  
  
  def __init__(self) -> None:
    pass
  
  
  def train(self,X, y):
    self.Xtr = X
    self.ytr = y
  

  def predict(self, X):
    X = np.array(X)
    num_test = X.shape[0]
    training_set_size = self.Xtr.shape[0]
    Y_prediction = np.zeros(shape=num_test, dtype=int)
    for i in range(num_test):
      Y_prediction[i] = self.ytr[np.argmin(np.sum(np.abs(self.Xtr - X[i,:]), axis = 1))]
    return Y_prediction    
```

```python
# data_batch_1.keys() : dict_keys([b'batch_label', b'labels', b'data', b'filenames'])
data_batch_1 = unpickle('cifar-10-python/cifar-10-batches-py/data_batch_1')
test_batch = unpickle('cifar-10-python/cifar-10-batches-py/test_batch')
X_train = data_batch_1[b'data']
y_train = data_batch_1[b'labels']
X_test = test_batch[b'data']
y_test = test_batch[b'labels']
```

```python
classifier = NearestNeighbor()
classifier.train(X=X_train,y=y_train)
test_result = classifier.predict(X_test)
n = (X_test.shape[0])
i = np.meshgrid(range(n))
accuracy = sum((test_result[i]==np.array(y_test)[i]))/n
print(f"[+] accuracy = {accuracy*100}%")
```
