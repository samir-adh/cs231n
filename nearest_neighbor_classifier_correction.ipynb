```python
path = "/content/drive/MyDrive/cs231n/"
def unpickle(file):
    import pickle
    with open(path + file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
class NearestNeighbor(object):
  def __init__(self):
    pass

  def train(self, X, y):
    """ X is N x D where each row is an example. Y is 1-dimension of size N """
    # the nearest neighbor classifier simply remembers all the training data
    self.Xtr = X
    self.ytr = y

  def predict(self, X):
    """ X is N x D where each row is an example we wish to predict label for """
    num_test = X.shape[0]
    # lets make sure that the output type matches the input type
    Ypred = np.zeros(num_test, dtype = self.ytr.dtype)

    # loop over all test rows
    for i in range(num_test):
      # find the nearest training image to the i'th test image
      # using the L1 distance (sum of absolute value differences)
      distances = np.sum(np.abs(self.Xtr - X[i,:]), axis = 1)
      min_index = np.argmin(distances) # get the index with smallest distance
      Ypred[i] = self.ytr[min_index] # predict the label of the nearest example
      print(f"{i/num_test * 100}%")

    return Ypred
```

```python
data_batch_1 = unpickle('cifar-10-python/cifar-10-batches-py/data_batch_1')
test_batch = unpickle('cifar-10-python/cifar-10-batches-py/test_batch')
X_train = np.array(data_batch_1[b'data'])
y_train = np.array(data_batch_1[b'labels'])
X_test = np.array(test_batch[b'data'])
y_test = np.array(test_batch[b'labels'])
```

```python
classifier = NearestNeighbor()
classifier.train(X=X_train,y=y_train)
test_result = classifier.predict(X_test)
n = range(X_test.shape[0])
i = np.meshgrid(n)
accuracy = sum(test_result[i][1]==y_test[i])/n
print(f"[+] accuracy = {accuracy*100}%")
```
